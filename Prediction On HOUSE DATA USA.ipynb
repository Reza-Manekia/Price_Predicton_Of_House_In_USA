## Splitting Of data set into train and test

```python
from sklearn.model_selection import train_test_split

dependants = ['sqft_living', 'bedrooms', 'bathrooms', 'view', 'yr_built']

X_train, X_test, y_train, y_test = train_test_split(df[dependants], df['price'], test_size=0.2, random_state=42)

print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

## Model Training (Linear Regression)  

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Calculating Mean_Absolute_error
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)

print("Mean Absolute Error:", mae)
```

```python
# Prediction of price of house when
# 2000 sq-ft 
# 3 bedrooms
# 2 bathrooms
# 0 view
# 1990 year_built 
```

```python
new_house_features = [2000, 3, 2, 0, 1990]

predicted_price = model.predict([new_house_features])

print("Predicted price:", round(predicted_price[0]))
```
